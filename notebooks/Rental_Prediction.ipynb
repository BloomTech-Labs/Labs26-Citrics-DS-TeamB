The modeling technique is used is called 'Facebook prophet'. Click [here](https://facebook.github.io/prophet/docs/quick_start.html) to learn more

At first I created a dataframe for the first city(City_id = 0) which includes predicted data for studio apartments only

```
df = pd.read_csv('rental_final.csv')
df1 = df.loc[df['city_id'] == 0][['year', 'fmr_0']]
df2 = df.loc[df['city_id'] == 0][['city_id', 'city', 'state', 'city_state', 'year', 'fmr_0']]
df1.columns = ['ds', 'y']
df1['ds'] = pd.to_datetime(df1['ds'], format='%Y')
m = Prophet()
m.fit(df1)
future0 = m.make_future_dataframe(periods=1827, freq= 'd')
print(future0.head(1))
future0['month'] = future0['ds'].dt.month
future0['day'] = future0['ds'].dt.day
future0 = future0.query('month==1 & day==1 ')
future = future0[['ds']]
forecast = m.predict(future)
df_studio = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
df_studio['year'] = forecast['ds'].dt.year
df_studio['city_id'] = df2['city_id'].iloc[0]
df_studio['city'] = df2['city'].iloc[0]
df_studio['state'] = df2['state'].iloc[0]
df_studio['city_state'] = df2['city_state'].iloc[0]
df_studio = df_studio[['city_id', 'city', 'state', 'city_state', 'ds', 'yhat', 'yhat_lower', 'yhat_upper', 'year']]
df_studio.to_csv('studio.csv', index=False)
df_studio.head(20)
```

```
import plotly.graph_objects as go
import numpy as np
def studio_rental_viz(df, city_id):

  df = df.loc[df['city_id'] == city_id]
  x = df['year'].tolist()
  x_rev = x[::-1]

  # Line 1
  y = df['st_yhat'].tolist()
  y_upper = df['st_yhat_upper'].tolist()
  y_lower = df['st_yhat_lower'].tolist()
  y_lower = y_lower[::-1]

  fig = go.Figure()

  fig.add_trace(go.Scatter(
      x=x+x_rev,
      y=y_upper+y_lower,
      fill='toself',
      fillcolor='rgba(0,100,80,0.2)',
      line_color='rgba(255,255,255,0)',
      showlegend=False,
      name='Studio',
  ))

  fig.add_trace(go.Scatter(
      x=x, y=y,
      line_color='rgb(0,100,80)',
      name='Studio',
  ))

  fig.update_traces(mode='lines')
  x = fig.show()
  return x
```
